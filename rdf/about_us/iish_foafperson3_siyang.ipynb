```python
#23-09-2015
#01-10-2015 modified
#06-10-2015 modified 1.delete vcard 2. delete Provenance

from rdflib import Graph, Namespace, RDF, Literal, RDFS, URIRef
from rdflib.namespace import FOAF

g = Graph()

PEOPLE = Namespace('http://qber.data2semantics.org/vocab/iisg/people/')
#IISG = Namespace('http://qber.data2semantics.org/vocab/iisg/')
CLARIAH = Namespace('http://qber.data2semantics.org/vocab/clariah/')
ORGANIZATION = Namespace('http://qber.data2semantics.org/vocab/clariah/organizations/')
ORG = Namespace('http://www.w3.org/ns/org#')
SKOS  = Namespace('http://www.w3.org/2004/02/skos/core#')

g.add((ORGANIZATION.IISG, RDF.type, FOAF.Agent))
g.add((ORGANIZATION.IISG, RDFS.subClassOf, ORG.Organization))
g.add((ORGANIZATION.IISG, SKOS.label, Literal("International Institute of Social History")))
g.add((ORGANIZATION.IISG, ORG.Site, Literal("Cruquiusweg 31 1019AT Amsterdam The Netherlands")))
g.add((ORGANIZATION.IISG, ORG.hasMember, PEOPLE.siyang))
g.add((ORGANIZATION.IISG, ORG.hasMember, PEOPLE.richard))
g.add((ORGANIZATION.IISG, FOAF.homepage, URIRef("https://socialhistory.org/")))

g.add((ORGANIZATION.VU, RDF.type, FOAF.Agent))
g.add((ORGANIZATION.VU, RDFS.subClassOf, ORG.Organization))
g.add((ORGANIZATION.VU, SKOS.label, Literal("Vrije Universiteit Amsterdam")))
g.add((ORGANIZATION.VU, ORG.hasMember, PEOPLE.kathrin))
g.add((ORGANIZATION.VU, ORG.hasMember, PEOPLE.rinke)) 

g.add((ORGANIZATION.UU, RDF.type, FOAF.Agent))
g.add((ORGANIZATION.UU, RDFS.subClassOf, ORG.Organization))
g.add((ORGANIZATION.UU, SKOS.label, Literal("Universiteit Utrecht")))
g.add((ORGANIZATION.UU, ORG.hasMember, PEOPLE.auke))
 
g.add((CLARIAH.CLARIAH, RDF.type, ORG.OrganizationalCollaboration)) 
g.add((CLARIAH.CLARIAH, FOAF.name, Literal("Common Lab Research Infrastructure for the Arts and Humanities")))      
g.add((CLARIAH.CLARIAH, FOAF.homepage, URIRef("http://www.clariah.nl/")))
g.add((CLARIAH.CLARIAH, ORG.hasMember, ORGANIZATION.IISG))#interpreting clariah as a group of members from different institutes.
g.add((CLARIAH.CLARIAH, ORG.hasMember, ORGANIZATION.VU))
g.add((CLARIAH.CLARIAH, ORG.hasMember, ORGANIZATION.UU))

g.add((CLARIAH.CSDH, RDF.type, ORG.OrganizationalUnit))
g.add((CLARIAH.CSDH, RDF.type, FOAF.Group))
g.add((CLARIAH.CSDH, FOAF.name, Literal("Clariah Structured Data Hub")))
g.add((CLARIAH.CSDH, FOAF.homepage, URIRef("https://clariah-sdh.github.io")))
g.add((CLARIAH.CSDH, ORG.unitOf, CLARIAH.CLARIAH))

g.add((PEOPLE.siyang, RDF.type, FOAF.Person))
g.add((PEOPLE.siyang, FOAF.surname, Literal("Kong")))
g.add((PEOPLE.siyang, FOAF.firstname, Literal("Siyang")))
g.add((PEOPLE.siyang, FOAF.homepage, URIRef("https://nl.linkedin.com/pub/siyang-kong/82/36b/610")))
g.add((PEOPLE.siyang, FOAF.mbox, URIRef("siyang.kong@iisg.nl")))
g.add((PEOPLE.siyang, FOAF.member, CLARIAH.CSDH ))


g.add((PEOPLE.kathrin, RDF.type, FOAF.Person))
g.add((PEOPLE.kathrin, FOAF.surname, Literal("Dentler")))
g.add((PEOPLE.kathrin, FOAF.firstname, Literal("Kathrin")))
g.add((PEOPLE.kathrin, FOAF.homepage, URIRef("https://nl.linkedin.com/in/dentler")))
g.add((PEOPLE.kathrin, FOAF.member, CLARIAH.CSDH ))

g.add((PEOPLE.rinke, RDF.type, FOAF.Person))
g.add((PEOPLE.rinke, FOAF.surname, Literal("Hoekstra")))
g.add((PEOPLE.rinke, FOAF.firstname, Literal("Rinke")))
g.add((PEOPLE.rinke, FOAF.homepage, URIRef("https://nl.linkedin.com/pub/rinke-hoekstra/2/9aa/28")))
g.add((PEOPLE.rinke, FOAF.member, CLARIAH.CSDH ))

g.add((PEOPLE.auke, RDF.type, FOAF.Person))
g.add((PEOPLE.auke, FOAF.surname, Literal("Rijpma")))
g.add((PEOPLE.auke, FOAF.firstname, Literal("Auke")))
g.add((PEOPLE.auke, FOAF.homepage, URIRef("https://nl.linkedin.com/pub/auke-rijpma/38/42b/128")))
g.add((PEOPLE.auke, FOAF.member, CLARIAH.CSDH ))

g.add((PEOPLE.richard, RDF.type, FOAF.Person))
g.add((PEOPLE.richard, FOAF.surname, Literal("Zijdeman")))
g.add((PEOPLE.richard, FOAF.firstname, Literal("Richard")))
g.add((PEOPLE.richard, FOAF.homepage, URIRef("https://nl.linkedin.com/in/richardzijdeman")))
g.add((PEOPLE.richard, FOAF.member, CLARIAH.CSDH ))

#g.bind('iisg', IISG)
g.bind('organization', ORGANIZATION)
g.bind('foaf', FOAF)
g.bind('clariah', CLARIAH)
g.bind('people', PEOPLE)
g.bind('org', ORG)
g.bind('skos', SKOS)

print  g.serialize(format='turtle')

with open('iish_foafperson3_siyang.ttl','w') as out:
     g.serialize(out, format='turtle')






```
